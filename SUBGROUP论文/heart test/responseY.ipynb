```python
import numpy as np
import matplotlib.pyplot as plt
import func
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.utils.fixes import parse_version
from sklearn.linear_model import LinearRegression
```

```python
fitdata=np.loadtxt('fitcleveland.csv',delimiter=',',skiprows=True)
```

```python
testdata=np.loadtxt('6cleveland.csv',delimiter=',',skiprows=True)
```

```python
Y=fitdata[:,0]
X=fitdata[:,1:]
```

```python
betaols=func.Beta0(X,Y)
betaols[0]
```

```python
fitY=np.dot(X,betaols[0])+betaols[1]
```

```python
np.savetxt('fitY.csv',fitY,delimiter=',')
```

```python
Xnew=testdata
```

```python
betaols0=func.Beta0(Xnew,fitY)
```

```python
betaols0
```

```python
ca0=(fitY-np.dot(Xnew,betaols0[0])).T
```

```python
ca=ca0[:,None]
```

```python
X_plot = np.linspace(140, 200, 1000)[:, np.newaxis]
fig, ax = plt.subplots(figsize=(8,10))
colors = ["navy", "cornflowerblue", "darkorange"]
kernels = ["gaussian", "tophat", "epanechnikov"]
lw = 2

for color, kernel in zip(colors, kernels):
    kde = KernelDensity(kernel=kernel, bandwidth=4).fit(ca)
    log_dens = kde.score_samples(X_plot)
    ax.plot(
        X_plot[:, 0],
        np.exp(log_dens),
        color=color,
        lw=lw,
        linestyle="-",
        label="kernel = '{0}'".format(kernel),
    )


ax.legend(loc="upper left")
ax.plot(ca[:, 0], -0.005 - 0.01 * np.random.random(ca.shape[0]), "+k")

plt.show()
```
