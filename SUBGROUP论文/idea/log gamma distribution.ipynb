```python
import numpy as np
from scipy.stats import gamma
from scipy.stats import beta
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import scipy
```

```python
fig = plt.figure(figsize=(15,5))


alpha_beta_values = np.array([[1,0.5], [2, 1], [3,1], [4,2], [5,3]])
linestyles = []
x = np.linspace(-5,5,2000) # 0150
xnum=np.exp(x)
y = np.linspace(-5,5,2000)

for alpha_beta_value in alpha_beta_values:
    print(alpha_beta_value)
    a=alpha_beta_value[0]
    b=np.exp(-scipy.special.digamma(a))
    y=((np.exp(-xnum/b))*((xnum)**(a)))/((scipy.special.gamma(a))*((b)**a))
    plt.plot(x, y, label=r'$\alpha=%.1f,\ \beta=%.1f$' % (alpha_beta_value[0], alpha_beta_value[1]))


#  x,y 
plt.xlim(-5, 5)
plt.ylim(0, 1)
plt.legend()
plt.show()
```
